## Tudor's Data

```python
freeboard = np.loadtxt("../data/CS2/TXTFILES/201304_CS2fb_1.5lon_x_0.5lat.txt", skiprows=1)
ak = np.loadtxt("../data/AK/TXTFILES/201304_AKfb_1.5lon_x_0.5lat.txt", skiprows=1)
freeboard = freeboard[~np.isnan(freeboard).any(axis=1)]
ak = ak[~np.isnan(ak).any(axis=1)]
```

```python
list(freeboard[70])
```

Where we have that each of the columns corresponds to:
- Longitude
- Latitude
- Freeboard measurement 
- Standard deviation
- Number of observations

```python
total_observations = freeboard.shape[0] + ak.shape[0]
#observations_cap = 500
observations_shape = (total_observations, 5)
observations_matrix = np.zeros(observations_shape)
```

```python
i = 0
for freeboard_row in freeboard:
    observations_matrix[i, 0] = freeboard_row[0] # Longitude
    observations_matrix[i, 1] = freeboard_row[1] # Latitude
    observations_matrix[i, 2] = 0                # Observation Category
    observations_matrix[i, 3] = freeboard_row[2] # Freeboard measurement (m)
    observations_matrix[i, 4] = freeboard_row[3] # Standard deviation 
    i += 1
for ak_row in ak:
    observations_matrix[i, 0] = ak_row[0] # Longitude
    observations_matrix[i, 1] = ak_row[1] # Latitude
    observations_matrix[i, 2] = 1       # Observation Category
    observations_matrix[i, 3] = ak_row[2] # Freeboard measurement (m)
    observations_matrix[i, 4] = ak_row[3] # Standard deviation 
    i += 1
```

```python
latitude_values = observations_matrix.T[0]
longitude_values = observations_matrix.T[1]
```

```python
print(f"MIN X: {min(longitude_values)}, MAX X: {max(longitude_values)}, MIN Y: {min(latitude_values)}, MAX Y: {max(latitude_values)}")
```

```python
np.savetxt("../snow_ice_observations_complete.txt", observations_matrix, '%5.1f %5.1f %d %5.5f %5.5f')

# Add the total number of observations at the top of the file
with open('../snow_ice_observations_complete.txt', 'r') as original: data = original.read()
with open('../snow_ice_observations_complete.txt', 'w') as modified: modified.write(f"{observations_matrix.shape[0]}\n" + data)
```

## Basic visualization using pandas

```python
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from itertools import chain
```

```python
data = pd.DataFrame(observations_matrix, columns=["Longitude", "Latitude", "Type", "Value", "StdDev"])
data_cs2 = data[data["Type"] == 0.0]
data_ak  = data[data["Type"] == 1.0]
```

### Orthogonal visualization in polar coordinates
First let's start by plotting the registered observations in the map to see roughly what shape we would expect.

```python
def draw_map(m, scale=0.2):
    # draw a shaded-relief image
    m.shadedrelief(scale=scale)
    
    # lats and longs are returned as a dictionary
    lats = m.drawparallels(np.linspace(-90, 90, 13))
    lons = m.drawmeridians(np.linspace(-180, 180, 13))

    # keys contain the plt.Line2D instances
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)
    
    # cycle through these lines and set the desired style
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')
```

```python
# 1. Draw map background:
plt.figure(figsize=(16, 8))
m = Basemap(projection='lcc', resolution=None, lat_0=-90, lon_0=0, lat_1=89.9, lon_1=180, width=1E7, height=0.5E7)
draw_map(m)

# 2. Scatter observation data
m.scatter(data_cs2['Longitude'], data_cs2['Latitude'], latlon=True, alpha=1, s=0.5)
```

### Select a subset of the data
In order to reduce the computational difficulty of the task, we can select only a subset of the data that we will actually pass into the inversion program.

```python
lat_lim = (72.5, 80)
lon_lim = (-180, -130)
```

```python
data_subset = data[
    (data["Latitude"] > lat_lim[0]) & (data["Latitude"] < lat_lim[1]) &
    (data["Longitude"] > lon_lim[0]) & (data["Longitude"] < lon_lim[1])]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
data_subset[data_subset['Type'] == 0.0].plot.scatter(ax= ax1, x='Longitude', y='Latitude', c='Value', cmap='seismic', title="CryptoSat-2")
data_subset[data_subset['Type'] == 1.0].plot.scatter(ax= ax2, x='Longitude', y='Latitude', c='Value', cmap='seismic', title="AltiKa")
```

```python
observations_matrix_subset = data_subset.values
```

```python
np.savetxt("../snow_ice_observations.txt", observations_matrix_subset, '%5.1f %5.1f %d %5.5f %5.5f')

# Add the total number of observations at the top of the file
with open('../snow_ice_observations.txt', 'r') as original: data = original.read()
with open('../snow_ice_observations.txt', 'w') as modified: modified.write(f"{observations_matrix_subset.shape[0]}\n" + data)
```

## Fill empty points

```python
MINLON=-180.0
MAXLON=178.5
MINLAT=60.0
MAXLAT=88.0
```

```python
longitudes = np.arange(MINLON, MAXLON+1.5, 1.5)
latitudes  = np.arange(MINLAT, MAXLAT+0.5, 0.5)
```

```python
data_cs2_fixed = data_cs2.copy()
data_ak_fixed = data_ak.copy()
```

```python
for i, lon in enumerate(longitudes):
    for j, lat in enumerate(latitudes):
        if len(data_cs2[(data_cs2.Longitude == lon) & (data_cs2.Latitude == lat)]) == 0:
            data_cs2_fixed = data_cs2_fixed.append({'Longitude': lon, 'Latitude': lat, 'Type': 0.0, 'Value': 0., 'StdDev': 0.5}, ignore_index=True)
        if len(data_ak[(data_ak.Longitude == lon) & (data_ak.Latitude == lat)]) == 0:
            data_ak_fixed = data_ak_fixed.append({'Longitude': lon, 'Latitude': lat, 'Type': 1.0, 'Value': 0., 'StdDev': 0.5}, ignore_index=True)
```

```python
data_cs2_fixed
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
data_cs2_fixed.plot.scatter(ax= ax1, x='Longitude', y='Latitude', c='Value', cmap='seismic', title="CryoSat-2")
data_ak_fixed.plot.scatter(ax= ax2, x='Longitude', y='Latitude', c='Value', cmap='seismic', title="AltiKa")

ax2.set_ylim(60, 90)
```

```python
fixed_data = data_cs2_fixed.append(data_ak_fixed)
```

```python
observations_matrix_subset = fixed_data.values
```

```python
np.savetxt("../snow_ice_observations_complete_fixed.txt", observations_matrix_subset, '%5.1f %5.1f %d %5.5f %5.5f')

# Add the total number of observations at the top of the file
with open('../snow_ice_observations_complete_fixed.txt', 'r') as original: data = original.read()
with open('../snow_ice_observations_complete_fixed.txt', 'w') as modified: modified.write(f"{observations_matrix_subset.shape[0]}\n" + data)
```
